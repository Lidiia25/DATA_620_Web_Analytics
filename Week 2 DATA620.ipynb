```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (10,8)
```

Create a graph object and add nodes

```python
g = nx.Graph()
```

```python
g.add_nodes_from(["Andre","Beverly","Ed","Carol","Diane","Fernando","Garth","Heather","Ike","Jane"])
```

Create edges

```python
g.add_edges_from([("Andre","Beverly"),("Andre","Carol"),("Andre","Diane"),("Andre","Fernando"),
                  ("Beverly","Diane"),("Beverly","Ed"),("Beverly","Garth"),
                  ("Carol","Diane"),("Carol","Fernando"),
                  ("Diane","Ed"),("Diane","Fernando"),("Diane","Garth"),
                  ("Ed","Garth"),
                  ("Fernando","Garth"),("Fernando","Heather"),
                  ("Garth","Heather"),
                  ("Heather","Ike"),
                  ("Ike","Jane")])
```

Draw the graph

```python
nx.draw(g, with_labels=True)
```
